### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

There is a significant difference in torque values between failed and non- failed machines.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Null Hypothesis(H0):**
There is no significant difference in torque values between faile dand non- failed machines.

**Alternative Hypothesis(H1):**There is significant difference in torque values between failed and non-faile dmachines.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import ttest_ind

failure = df[df['Machine failure']==1]['Torque [Nm]']
no_failure = df[df['Machine failure']==0]['Torque [Nm]']

t_stat, p_value = ttest_ind(failure, no_failure)

print("T-Statistic:", t_stat)
print("P-value:", p_value)
````

##### Which statistical test have you done to obtain P-Value?

An Independent two simple T-Test was performed to obtain teh p_value.

##### Why did you choose the specific statistical test?

 The independent two simple T_test was chosen because the objective is to compare the mean torque values of two independent groups: failed machines and non-failed machines.Since torque si a numerical variable and machine failure is a binary categorical variable, the t-test is appropriate for determining whether a significant difference exists between the two groups.

#### 2. Cross- Validation & Hyperparameter Tuning

````python
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder, StandardScaler
from imblearn.over_sampling import SMOTE  # Imbalance handle karne ke liye

# --- 1. Dataset Loading ---
df = pd.read_csv('/content/train.csv')

# --- 2. Drop Unnecessary Columns ---
df.drop(['id', 'Product ID'], axis=1, inplace=True)

# --- 3. Categorical Encoding ---
le = LabelEncoder()
df['Type'] = le.fit_transform(df['Type'])

# --- 4. Feature Engineering ---
df['temp_Difference'] = df['Process temperature [K]'] - df['Air temperature [K]']
df['Power_Proxy'] = df['Torque [Nm]'] * df['Rotational speed [rpm]']

# --- 5. Data Transformation (Log Transform) ---
df['Tool wear [min]'] = np.log1p(df['Tool wear [min]'])

# --- 6. Data Scaling (Sabhi numerical features ko scale kiya) ---
scaler = StandardScaler()
columns_to_scale = [
    'Torque [Nm]', 'Rotational speed [rpm]', 'Tool wear [min]',
    'Process temperature [K]', 'Air temperature [K]',
    'temp_Difference', 'Power_Proxy'  # Naye features ko bhi add kiya
]
df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])

# --- 7. Data Splitting ---
x = df.drop(columns=['Machine failure'])
y = df['Machine failure']
# Data leakage se bachne ke liye stratify=y lagaya taaki train/test dono me failure ratio sahi rahe
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)

# --- 8. Handling Imbalanced Dataset ---
smote = SMOTE(random_state=42)
x_train_resampled, y_train_resampled = smote.fit_resample(x_train, y_train)

# --- 9. Hyperparameters & GridSearchCV ---
param_grid = {
    'C': [0.01, 0.1, 1, 10],
    'solver': ['liblinear', 'lbfgs']
}

# max_iter=1000 kiya taaki model bina warning ke puri tarah converge (sahi se solve) ho sake
lr_model = LogisticRegression(random_state=42, max_iter=1000)

# Imbalanced data predictive maintenance ke liye 'f1' ya 'roc_auc' scoring zyada sahi hoti hai accuracy ke mukable
grid_search = GridSearchCV(lr_model, param_grid, cv=5, scoring='f1', n_jobs=-1)

# --- 10. Fit the Algorithm ---
grid_search.fit(x_train_resampled, y_train_resampled)

# --- 11. Evaluation ---
best_model = grid_search.best_estimator_
y_pred_tuned = best_model.predict(x_test)

print("Best Parameters:", grid_search.best_params_)
print("Best Cross Validation Score (F1-Score):", grid_search.best_score_)
````

##### Which hyperparameter optimization technique have you used and why?

I used GridSearchCV for hyperparameter optimization. It tests multiple combinations of hyperparameters and selects the best_performing model using cross_validation. This helps improve model performance and reduces the risk of overfitting.

##### Have you seen any improvement? Note down the improvement with updates Evaluation metric Score Chart.

Yes, after applying GridSearchCV, the model performance improved compared to the default Logistic Regression model. The optimized model achieved better prediction accuracy and more reliable classification results.
